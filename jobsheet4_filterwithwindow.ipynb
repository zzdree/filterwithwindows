```python
# Input parameters
N = 256
NFFT = 2048

# Input signal (long: N >> NFFT)
tt = np.linspace(0, 1, N, endpoint=True)
xx = np.cos(2*np.pi*32*tt)

# Calculate FFT
sft = np.abs(fft(xx, NFFT))
slg = 20*np.log10(sft / np.max(sft))

plt.figure(figsize=(16, 6), dpi=80)
plt.subplot(2,1,1)
plt.title('Periodic signal, N = {}'.format(N))
plt.xlabel('time samples')
plt.ylabel('Magnitude')
plt.plot(xx)
plt.xlim([0, N-1])
plt.grid(True)

plt.subplot(2,1,2)
plt.title('Spectrum, NFFT = {}'.format(NFFT))
plt.xlabel('freq samples')
plt.ylabel('level [dB]')
plt.plot(slg)
plt.xlim([0, NFFT//2])
plt.ylim([-40, 0])
plt.grid(True)
plt.tight_layout()
```

Program ini membuat sinyal periodik dengan frekuensi 32 Hz dengan jumlah sampel yang lebih besar (N = 65536), menghitung FFT-nya, dan memvisualisasikan sinyal periodik dan spektrumnya dalam dua subplot yang terpisah. Subplot pertama menampilkan sinyal periodik dalam domain waktu, sedangkan subplot kedua menampilkan spektrum sinyal dalam domain frekuensi (dengan skala dB) dengan rentang sumbu x dan y yang berbeda dari kode sebelumnya.

```python
def dft_win(w):
    """
    Plot window function and spectrum of the window

    Parameters
    ----------
    w : np.array
        input window vector
    """

    # Find length of window
    N = len(w)
    # FFT size
    NFFT = 2**12
    # Calculate FFT
    W = fftshift(fft(w, NFFT))
    # Find max and add minimum floating-point value (except divide-by-zero)
    W = W / np.amax(W) + np.nextafter(0,1)
    tt = np.linspace(-1, 1, NFFT)

    # plot window function and its spectrum
    fig = plt.figure(figsize=(16, 6))

    plt.subplot(2,1,1)
    plt.stem(w, use_line_collection=True, basefmt='C0')
    plt.title('Window fucntion')
    plt.xlabel('Samples')
    plt.ylabel(r'$w[k]$')
    plt.xlim([0, N-1])
    plt.grid(True)

    plt.subplot(2,1,2)
    plt.plot(tt, 20*np.log10(np.abs(W)))
    plt.title('Spectrum')
    plt.xlabel('Frequency')
    plt.ylabel('Level in dB')
    plt.axis([-1, 1, -120, 2])
    plt.grid(True)
    plt.tight_layout()
```

Source code tersebut merupakan sebuah fungsi Python bernama dft_win yang digunakan untuk memvisualisasikan sebuah fungsi window (window function) dan spektrumnya menggunakan Transformasi Fourier Cepat (Fast Fourier Transform/FFT).

1.	Fungsi dft_win menerima satu parameter yaitu w yang merupakan vektor window (window vector) dalam bentuk array NumPy.
2.	N = len(w) mendapatkan panjang dari vektor window w.
3.	NFFT = 2**12 mendefinisikan jumlah sampel FFT yang digunakan, dalam hal ini 2^12 = 4096.
4.	W = fftshift(fft(w, NFFT)) menghitung FFT dari vektor window w dengan jumlah sampel NFFT, dan kemudian melakukan operasi fftshift untuk memindahkan komponen frekuensi nol ke tengah spektrum.
5.	W = W / np.amax(W) + np.nextafter(0,1) melakukan normalisasi terhadap hasil FFT W dengan membaginya dengan nilai maksimum dan menambahkan nilai floating-point terkecil berikutnya untuk menghindari pembagian dengan nol.
6.	tt = np.linspace(-1, 1, NFFT) membuat vektor yang berisi nilai-nilai frekuensi yang terdistribusi secara merata antara -1 dan 1 dengan jumlah sampel NFFT.
7.	fig = plt.figure(figsize=(16, 6)) membuat figure baru dengan ukuran tertentu.
8.	plt.subplot(2,1,1) membuat subplot pertama (atas) dengan tata letak 2 baris dan 1 kolom.
9.	plt.stem(w, use_line_collection=True, basefmt='C0') memplot vektor window w menggunakan fungsi stem yang menampilkan garis vertikal dari sumbu x ke setiap nilai data.
10.	plt.title('Window fucntion'), plt.xlabel('Samples'), plt.ylabel(r'$w[k]$') memberikan judul dan label sumbu untuk subplot pertama.
11.	plt.xlim([0, N-1]) dan plt.grid(True) mengatur batas sumbu x dan menampilkan grid pada subplot.
12.	plt.subplot(2,1,2) membuat subplot kedua (bawah) dengan tata letak 2 baris dan 1 kolom.
13.	plt.plot(tt, 20*np.log10(np.abs(W))) memplot spektrum window dalam skala desibel (dB) dengan menggunakan vektor frekuensi tt dan hasil FFT W.
14.	plt.title('Spectrum'), plt.xlabel('Frequency'), dan plt.ylabel('Level in dB') memberikan judul dan label sumbu untuk subplot kedua.
15.	plt.axis([-1, 1, -120, 2]) mengatur batas sumbu x dari -1 hingga 1 dan batas sumbu y dari -120 dB hingga 2 dB.
16.	plt.grid(True) dan plt.tight_layout() menampilkan grid pada subplot dan menyesuaikan tata letak figure secara keseluruhan.

Fungsi dft_win ini dapat digunakan untuk memvisualisasikan fungsi window dan spektrumnya dalam dua subplot yang terpisah. ```python

```python
N = 64
# All windows
ww = np.zeros((N, 9))

ww[:,0] = np.ones(N)
ww[:,1] = sig.triang(N)
ww[:,2] = sig.hann(N)
ww[:,3] = sig.hamming(N)
ww[:,4] = sig.blackman(N)
ww[:,5] = sig.blackmanharris(N)
ww[:,6] = sig.nuttall(N)
ww[:,7] = sig.flattop(N)
ww[:,8] = sig.kaiser(N, beta=8)

lst_titles = ['Rectangular', 'Bartlett', 'Hann (Hanning)',
              'Hamming', 'Blackman', 'Blackmann-Harris 4-term',
              'Nuttall', 'Flat-top window', 'Kaiser (w/ beta)'
             ]

# Plot window function and its spectrum
fig = plt.figure(figsize=(16, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.stem(ww[:,i], use_line_collection=True, basefmt='C0')
    plt.title(lst_titles[i])
    plt.xlabel('Samples')
    plt.ylabel(r'$w[k]$')
    plt.xlim([0, N-1])
    plt.grid(True)
plt.tight_layout()
```

Source code ini membuat plot dari 9 fungsi window yang umum digunakan dalam pemrosesan sinyal dan menampilkan spektrumnya. Fungsi window tersebut adalah Rectangular, Bartlett, Hann, Hamming, Blackman, Blackman-Harris, Nuttall, Flat-top, dan Kaiser.

```python

NFFT = 2**12
# Calculate FFT
WW = fft(ww, NFFT, axis=0)
WW = WW / np.amax(WW, axis=0) + np.nextafter(0,1)

tt = np.linspace(-1, 1, NFFT)

# Plot window function and its spectrum
fig2 = plt.figure(figsize=(16, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.plot(tt, 20*np.log10(np.abs(fftshift(WW[:,i]))))
    plt.title(lst_titles[i])
    plt.xlabel('Frequency')
    plt.ylabel('Level in dB')
    plt.axis([-1, 1, -120, 2])
    plt.grid(True)
plt.tight_layout()
```

Source code ini melanjutkan code sebelumnya untuk menghitung dan menampilkan spektrum dari masing-masing fungsi window yang telah dibuat. Spektrum ini menunjukkan distribusi frekuensi yang terkandung dalam fungsi window.

1.	Konfigurasi FFT:
•	NFFT: Variabel yang menyimpan jumlah titik yang digunakan untuk perhitungan Fast Fourier Transform (FFT). Biasanya dipilih sebagai pangkat 2 untuk efisiensi komputasi (diset sebagai 2 pangkat 12).
2.	Perhitungan FFT:
•	WW: Array yang sebelumnya menyimpan nilai window, digunakan kembali untuk menyimpan hasil perhitungan FFT.
•	fft(ww, NFFT, axis=0): Fungsi fft dari library np digunakan untuk menghitung FFT dari setiap kolom (sesuai axis=0) pada array ww.
o	Dengan ini, spektrum untuk setiap fungsi window akan disimpan pada baris yang berbeda dalam array WW.
•	WW / np.amax(WW, axis=0): Normalisasi dilakukan terhadap nilai maksimum pada setiap kolom dari WW untuk memastikan skala yang wajar pada plot spektrum nanti.
•	+ np.nextafter(0,1): Penambahan nilai epsilon yang sangat kecil membantu menghindari pembagian dengan nol pada saat logaritma diambil nanti.
3.	Skala frekuensi:
•	tt: Array yang menyimpan nilai frekuensi normalized antara -1 dan 1 dengan jumlah titik NFFT.
o	Nilai frekuensi ini didapatkan menggunakan fungsi np.linspace.
4.	Membuat plot spektrum:
•	fig2: Variabel untuk menyimpan objek figure baru yang akan menampilkan plot spektrum.
•	Looping dilakukan untuk membuat 9 subplot pada figure tersebut (sama seperti sebelumnya).
o	Setiap iterasi looping (i):
	Subplot ke i+1 dipilih.
	fftshift(WW[:,i]): Fungsi untuk memusatkan komponen frekuensi DC (Direct Current) ke tengah plot.
	np.abs(): Fungsi untuk mengambil nilai absolut dari hasil fftshift.
	20*np.log10(): Fungsi untuk menghitung nilai spektrum dalam satuan dB.
	plt.plot(tt, ...): Fungsi untuk menampilkan plot spektrum dari fungsi window ke-i.
	Judul plot, label sumbu, dan batas sumbu diatur.
	Grid pada plot ditampilkan.
•	plt.tight_layout(): Fungsi untuk mengatur jarak antar subplot agar tidak saling tumpang tindih.

```python
# Input parameters
N = 128
NFFT = 2048

# Input signal (long: N >> NFFT)
tt = np.linspace(0, 1, N, endpoint=True)
xx = 100*np.cos(2*np.pi*32*tt)+np.cos(2*np.pi*40*tt)

# Window (Kaiser)
wn = sig.kaiser(N, beta=7)
yy = xx*wn

# Calculate FFT 1
sft = np.abs(fft(xx, NFFT))
slg = 20*np.log10(sft / np.max(sft))

# Calculate FFT 2
yft = np.abs(fft(yy, NFFT))
ylg = 20*np.log10(yft / np.max(yft))

lst_ffts = [slg, ylg]
lst_wins = ['No window', 'Kaiser window']
plt.figure(figsize=(16, 5))
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.title(lst_wins[i])
    plt.xlabel('freq samples')
    plt.ylabel('level [dB]')
    plt.plot(lst_ffts[i])
    plt.xlim([0, NFFT//2])
    plt.ylim([-80, 0])
    plt.grid(True)
plt.tight_layout()
```

Source code ini mendemonstrasikan pengaruh penggunaan fungsi window (Kaiser) terhadap spektrum sinyal menggunakan FFT (Fast Fourier Transform). Sinyal yang digunakan terdiri dari dua sinusoida dengan frekuensi berbeda.

1.	Konfigurasi parameter:
•	N: Jumlah sampel sinyal input (diset menjadi 128).
•	NFFT: Jumlah titik yang digunakan untuk perhitungan FFT (biasanya dipilih sebagai pangkat 2 untuk efisiensi komputasi, diset menjadi 2048).
2.	Sinyal input:
•	tt: Array yang menyimpan nilai waktu normalized antara 0 dan 1 dengan jumlah titik N.
•	xx: Array yang menyimpan sinyal input berupa penjumlahan dua sinusoida dengan frekuensi 32 Hz dan 40 Hz. Amplitudo sinusoida pertama adalah 100 kali lebih besar dari sinusoida kedua.
3.	Fungsi window (Kaiser):
•	wn: Array yang menyimpan nilai fungsi window Kaiser dengan jumlah titik N dan parameter beta 7.
o	Fungsi sig.kaiser dari library SciPy digunakan untuk mendapatkan window Kaiser.
4.	Sinyal dengan window:
•	yy: Array yang menyimpan sinyal setelah dikalikan dengan fungsi window Kaiser element-wise.
5.	Perhitungan FFT:
•	sft: Array yang menyimpan nilai spektrum magnitudo (absolut) dari sinyal input xx menggunakan perhitungan FFT.
o	Fungsi fft dari library NumPy digunakan untuk menghitung FFT.
o	NFFT digunakan sebagai jumlah titik untuk FFT.
•	slg: Array yang menyimpan nilai spektrum dalam satuan dB setelah dinormalisasi terhadap nilai maksimum pada sft.
o	20*np.log10( ... ) digunakan untuk konversi ke dB.
•	yft: Array yang menyimpan nilai spektrum magnitudo dari sinyal dengan window yy menggunakan perhitungan FFT (proses serupa dengan sft).
•	ylg: Array yang menyimpan nilai spektrum dalam satuan dB setelah dinormalisasi terhadap nilai maksimum pada yft.
6.	Visualisasi:
•	Dua buah subplot dibuat pada sebuah figure untuk menampilkan perbandingan spektrum.
•	lst_ffts: List yang menyimpan array slg (spektrum tanpa window) dan ylg (spektrum dengan window Kaiser).
•	lst_wins: List berisi label untuk setiap plot (tanpa window dan window Kaiser).
•	Setiap subplot:
o	Judul plot diatur menggunakan list lst_wins.
o	Label sumbu X dan Y diatur.
o	Plot spektrum ditampilkan menggunakan data dari lst_ffts.
o	Batas sumbu X dan Y diatur untuk fokus pada bagian penting dari spektrum.
o	Grid pada plot ditampilkan.
•	plt.tight_layout(): Fungsi untuk mengatur jarak antar subplot agar tidak saling tumpang tindih.
